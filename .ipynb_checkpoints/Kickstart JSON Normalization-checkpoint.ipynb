```python
import pandas as pd
import json
```

```python
data=pd.read_csv('Kickstarter057.csv')
data.head()
```

```python
list(data.columns)
```

```python
#Prima di tutto leggiamo la colonna e la convertiamo da Json ad un DataFrame
normalized_category=pd.json_normalize(data.category.apply(json.loads))

#Cambiamo i nomi delle colonne in modo da distinguere che si tratti di colonne derivanti da "Category"
normalized_category.columns=normalized_category.columns.map(lambda x: str(x) + '_category')
normalized_category
```

```python
#Adesso fondiamo insieme i data originali con il nuovo dataframe che contiene le colonne di "Category". Facciamo una inner JOIN
merged_data = pd.merge(data, normalized_category, left_index=True, right_index=True)
#Rimuoviamo la colonna originale "Category"
del merged_data['category']
merged_data
```

## Colonna Creator

Il procedimento Ã¨ lo stesso di prima solo che qua ho visto che ci sono dei JSON che hanno erroi (caratteri strani e doppi apici) in questi casi metto tutta la riga a NaN (cioÃ¨ un JSON vuoto).
Si puÃ² fare la stessa cosa anche sopra e sostituire il metodo json.loads con questa funzione sotto.

```python
def besteffort_loads(obj):
    try:
        return json.loads(obj)
    except (TypeError, ValueError):
        return {}
```

```python
#Dal Momento che ci sono dei JSON con errori, uso una funziona (besteffort_loads) che in caso di errori restituisce il JSON vuoto
normalized_creator=pd.json_normalize(data.creator.apply(besteffort_loads))

#Cambiamo i nomi delle colonne in modo da distinguere che si tratti di colonne derivanti da "Category"
normalized_creator.columns=normalized_creator.columns.map(lambda x: str(x) + '_creator')
normalized_creator
```

```python
#Adesso fondiamo insieme i data originali con il nuovo dataframe che contiene le colonne di "Category". Facciamo una inner JOIN
merged_data = pd.merge(merged_data, normalized_creator, left_index=True, right_index=True)
#Rimuoviamo la colonna originale "creator"
del merged_data['creator']
merged_data
```
